```python
# Import necessary libraries for data manipulation and numerical operations
import pandas as pd
import numpy as np
```

```python
# Define the model and data type for the labeling process
model = "UCD"  # The classifier for which labels are created (RCL/UCD/blink)
data = "UCD"   # The type of movement on which data is collected (RCL/UCD/blink)

# Define the file name where the data is stored
file = "file.csv"  # The CSV file containing the data

# Read the data from the CSV file into a pandas DataFrame
df = pd.read_csv(file)

# Define the duration up to which the data should be processed
duration = 120  # The maximum timestamp to include in the processing
```

```python
# Function to update labels based on specific conditions
def update_label(row):
    timestamp = row['Timestamp']  # Extract the 'Timestamp' value from the row

    # Check if the model is "blink"
    if model == "blink":
        # Check if the data is also "blink"
        if data == "blink":
            # This is the duration in which the subject blinks
            if timestamp % 2 >= 0.1 and timestamp % 2 < 0.9:
                return 1  # Label for blink
            else:
                return 0  # Label for no blink
        else:
            return 0  # Label for no blink
    else:
        # Check if the model and data match
        if model == data:
            # This is the duration when the subject moves their eye
            if timestamp % 2 >= 0.15 and timestamp % 2 < 0.55:
                if (timestamp // 2) % 4 == 3:
                    return 2  # Rightward/Upward movement
                elif (timestamp // 2) % 4 == 0:
                    return 0  # Leftward/Downward movement
                elif (timestamp // 2) % 4 == 1:
                    return 0  # Leftward/Downward movement 
                elif (timestamp // 2) % 4 == 2:
                    return 2  # Rightward/Upward movement
            else:
                return 1  # No movement
        else:
            return 1  # No movement
```

```python
# Filter the DataFrame to include data collected properly
df = df[df['Timestamp'] <= duration]

# Apply the update_label function to each row and create a new 'Label' column
# The lambda function passes each row to the update_label function
df['Label'] = df.apply(lambda row: update_label(row), axis=1)

# Save the updated DataFrame with the new 'Label' column to a CSV file
df.to_csv(file)
```
